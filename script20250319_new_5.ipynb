```python
import matplotlib.pyplot as plt
import pandas as pd


total_record = pd.read_csv('scripts/5_False.csv', index_col=0)
total_record = total_record.loc[total_record['iter'] == 40000]

plt.figure(figsize=(6, 6))

cmap = plt.get_cmap('coolwarm') 
norm = plt.Normalize(vmin=0, vmax=len(total_record.groupby('dz')) - 1) 

counter = 0
for dz, group in total_record.groupby('dz'):
    if dz <= 8:
        # 按dz排序确保连线正确
        sorted_group = group.sort_values('chi')
        plt.plot(sorted_group['chi'], - sorted_group['act_mse'] + sorted_group['obs_mse'] + sorted_group['noi_mse'], 
                label=f'$d_zs$={dz} LLO', marker='o', markersize=4, color=cmap(norm(counter)))
        # plt.plot(sorted_group['chi'], sorted_group['act_mse'], label=f'  Action MSE', ls=':', marker='^', markersize=4, color=cmap(norm(counter)))
        # plt.plot(sorted_group['chi'], sorted_group['noi_mse'] * (sig_value ** 2), label=f'  Noise MSE', ls=':', marker='x', markersize=4, color=cmap(norm(counter)))
        counter += 1

plt.plot([0, 17], [2, 2], 'k:', label='Optimal')

plt.xlim([0.0, 1.0])
plt.title(r'Linear LAM Objective', fontsize=13)
plt.xlabel('Bottleneck size $d_z$', fontsize=13)
plt.legend(fontsize=13)
```
